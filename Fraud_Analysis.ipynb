```python
Q3 = np.percentile(data['Amount'], 75)
Q1 = np.percentile(data['Amount'], 25)

cutoff = 5.0

IQR = (Q3 - Q1)

lower_bound = Q1 - (IQR * cutoff)
upper_bound = Q3 + (IQR * cutoff)

filter_data = (data['Amount'] < lower_bound) | (data['Amount'] > upper_bound)

outliers = data[filter_data]['Amount']
fraud_outliers = data[(data['Class'] == 1) & filter_data]['Amount']
normal_outliers = data[(data['Class'] == 0) & filter_data]['Amount']

print(f"Total Number of Outliers : {outliers.count()}")
print(f"Number of Outliers in Fraudulent Class : {fraud_outliers.count()}")
print(f"No of Outliers in Normal Class : {normal_outliers.count()}")
print(f"Percentage of Fraud amount outliers : {round((fraud_outliers.count()/outliers.count())*100,2)}%")
```

```python
import statsmodels.api as sm

def vif(df, col_index):
    vif = [1]
    for i in range(len(df.columns)):
        if i != col_index:
            x = df.iloc[:, i]
            y = df.iloc[:, col_index]
            r = sm.OLS(endog=y, exog=x).fit().rsquared
            vif.append(1 / (1 - r))
    return vif[col_index]

vif_df = pd.DataFrame()
vif_df["features"] = data.columns
vif_df["vif_value"] = [vif(data, i) for i in range(len(data.columns))]

print(vif_df)
```

```python
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler()
data['Time'] = robust_scaler.fit_transform(data['Time'].values.reshape(-1,1))
```

```python
X = data.drop(['Class'],axis=1)
Y = data['Class']
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE


print(f'Original dataset shape : {Counter(Y)}')

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, Y)

print(f'Resampled dataset shape {Counter(y_res)}')
```

```python
from sklearn.model_selection import StratifiedKFold
strat = StratifiedKFold(n_splits=5, random_state=None, shuffle=False)

for train_index, test_index in strat.split(X_res, y_res):
    print("Train:", train_index, "Test:", test_index)
    original_Xtrain, original_Xtest = X_res.iloc[train_index], X_res.iloc[test_index]
    original_ytrain, original_ytest = y_res.iloc[train_index], y_res.iloc[test_index]
```

```python
original_Xtrain = original_Xtrain.values
original_Xtest = original_Xtest.values
original_ytrain = original_ytrain.values
original_ytest = original_ytest.values
```

```python

from sklearn.linear_model import LogisticRegression

model_log = LogisticRegression()

param_grid_log = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear']
}

grid_log = RandomizedSearchCV(model_log, param_grid_log, cv=5, scoring='recall', n_jobs=1)

grid_log.fit(original_Xtrain, original_ytrain)

pred_log = grid_log.predict(original_Xtest)

print(f"Train_accuracy: {grid_log.score(original_Xtrain, original_ytrain)*100:.2f}%")
print(f"Test_accuracy: {grid_log.score(original_Xtest, original_ytest)*100:.2f}%")
print(f"accuracy_score: {accuracy_score(pred_log, original_ytest)*100:.2f}%")
print(f"roc_auc_score: {roc_auc_score(original_ytest, grid_log.predict_proba(original_Xtest)[:,1])*100:.2f}%")
print(f"precision_score: {precision_score(pred_log, original_ytest)*100:.2f}%")
print(f"recall_score: {recall_score(pred_log, original_ytest)*100:.2f}%")
print(f"f1_score: {f1_score(pred_log, original_ytest)*100:.2f}%")
show_metrics(grid_log, original_ytest, pred_log)
```

```python
coefs = grid_log.best_estimator_.coef_
importance = abs(coefs).sum(axis=0)
features = data.drop(['Class'], axis=1).columns
importance_df = pd.DataFrame({'feature': features, 'importance': importance})
importance_df = importance_df.sort_values(by='importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=importance_df)
plt.title('Feature Importance', fontsize=15)
plt.show()
```
